```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# reading from csv file
df = pd.read_csv("Data/Social_Network_Ads.csv")
df.head()
```

```python
# dropping user ID column, mapping gender Male->1, and Female->0
df.drop('User ID', axis=1, inplace=True)
df['Gender'] = df['Gender'].map({'Male':1, 'Female':0})
df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
# splitting df into features and target data
# and also splitting data into training and testing
x = df.drop('Purchased', axis=1)
y = df['Purchased']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=38)
```

```python
# Feature Scaling using StandardScaler
scaler = StandardScaler()
p1_x_train = scaler.fit_transform(x_train)
p1_x_test = scaler.transform(x_test)
```

```python
p1_x_train = pd.DataFrame(data=p1_x_train, columns=x_train.columns)
p1_x_test = pd.DataFrame(data=p1_x_test, columns=x_test.columns)
```

```python
np.round(p1_x_train.describe(), 2)
```

```python
fig,(X1, X2) = plt.subplots(1, 2, figsize=(15, 4))
plt.title("No change in distribution after Feature Scaling")
X1.scatter(x_train['Age'], x_train['EstimatedSalary'])
X2.scatter(p1_x_train['Age'], p1_x_train['EstimatedSalary'])
plt.xlabel("p1_Age")
plt.ylabel("p1_EstimatedSalary")
plt.show()
```
