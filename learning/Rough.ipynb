```python
import numpy as np
import pandas as pd
```

```python
data = [['2020-12-8', 'toyota', 0, 1], ['2020-12-8', 'toyota', 1, 0], ['2020-12-8', 'toyota', 1, 2], ['2020-12-7', 'toyota', 0, 2], ['2020-12-7', 'toyota', 0, 1], ['2020-12-8', 'honda', 1, 2], ['2020-12-8', 'honda', 2, 1], ['2020-12-7', 'honda', 0, 1], ['2020-12-7', 'honda', 1, 2], ['2020-12-7', 'honda', 2, 1]]
df = pd.DataFrame(data, columns=['date_id', 'make_name', 'lead_id', 'partner_id']).astype({'date_id':'datetime64[ns]', 'make_name':'object', 'lead_id':'Int64', 'partner_id':'Int64'})
```

```python
df.head()
```

```python
groupdf = df.groupby(['date_id', 'make_name']).agg({'lead_id':'nunique', 'partner_id':'nunique'}).reset_index()
groupdf
```

```python
groupdf.rename(columns={'lead_id':'unique_leads', 'partner_id':'unique_partners'}, inplace=True)
groupdf
```

```python
data = [[101, 'John', 'A', None], [102, 'Dan', 'A', 100], [103, 'James', 'A', 100], [104, 'Amy', 'A', 100], [105, 'Anne', 'A', 100], [106, 'Ron', 'B', 100], [107, 'Ron', 'B', 100]]
employee = pd.DataFrame(data, columns=['id', 'name', 'department', 'managerId']).astype({'id':'Int64', 'name':'object', 'department':'object', 'managerId':'Int64'})
```

```python
employee
```

```python
count = employee['managerId'].value_counts().reset_index(name='cnt')
count_5 = count[(count['cnt']>=5)]
count_5
```

```python
joindf = pd.merge(count_5, employee, how='inner', left_on='managerId', right_on='id')
joindf
```

```python
joindf.dropna(subset=['id'], axis=0, how='any', inplace=True)
joindf[['name']]
```

```python
data = [[1, '2017-01-01', 10], [2, '2017-01-02', 109], [3, '2017-01-03', 150], [4, '2017-01-04', 99], [5, '2017-01-05', 145], [6, '2017-01-06', 1455], [7, '2017-01-07', 199], [8, '2017-01-09', 188]]
df = pd.DataFrame(data, columns=['id', 'visit_date', 'people']).astype({'id':'Int64', 'visit_date':'datetime64[ns]', 'people':'Int64'})
```

```python
df.head()
```

```python
df.sort_values(by='id', inplace=True)
people_limit = df[(df['people']>=100)]
people_limit.head()
```

```python
group_key = (people_limit['id'].diff() != 1).cumsum()
```

```python
filtered_df = people_limit.groupby(group_key).filter(lambda g: len(g) > 3)
filtered_df
```
