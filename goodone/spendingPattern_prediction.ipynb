# Combining both datas

```python
df2.rename(columns={'major':'major_sub', 'shoping':'shopping'}, inplace=True)
df = pd.concat([df1, df2], ignore_index=True)
df.shape
```

```python
df.head()
```

# Data preProcessing

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
# OneHotEncoding
ohe = OneHotEncoder(sparse_output=False)
temp_data = ohe.fit_transform(df[['major_sub']])
tempdf = pd.DataFrame(data=temp_data, columns=ohe.get_feature_names_out())
tempdf.head()
```

```python
finaldf = pd.concat([tempdf, df], axis=1)
finaldf = finaldf.drop('major_sub', axis=1)
finaldf.head()
```

```python
finaldf.info()
```

```python
# splitting df into features and target data
# and also splitting data into training and testing
x = finaldf.drop('spent', axis=1)
y = finaldf['spent']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
x_train.info()
```

```python
# Feature Scaling using StandardScaler
scaler = StandardScaler()
p1_x_train = scaler.fit_transform(x_train)
p1_x_test = scaler.transform(x_test)
```

```python
# Feature Scaling using StandardScaler
p1_y_train = scaler.fit_transform(y_train.values.reshape(-1, 1))
p1_y_test = scaler.transform(y_test.values.reshape(-1, 1))
```

# Model Training

```python
# Training model
lr = LinearRegression()
lr.fit(p1_x_train, p1_y_train)
```

```python
# Getting predictions to check for model accuracy
y_pred = lr.predict(p1_x_test)
```

```python
# Accuracy metrics: R^2 Score, Mean Absolute Error (MAE), Mean Squared Error (MSE) and Root Mean Squared Error (RMSE)(not done)
R2 = r2_score(p1_y_test, y_pred)
MAE = mean_absolute_error(p1_y_test, y_pred)
MSE = mean_squared_error(p1_y_test, y_pred)
```

```python
print(R2)
print(MAE)
print(MSE)
```

```python
# mean absolute error graph
plt.scatter(p1_y_test,y_pred)
plt.xlabel('actual values')
plt.ylabel('predicted values')
plt.title('Actual vs Predicted')
plt.savefig('Fig3.png', dpi=300, bbox_inches='tight')
plt.show()
```
