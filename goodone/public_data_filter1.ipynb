```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../learning/Data/public_spending_data.csv")
df.head()
```

```python
# 1. Identify the rows that have 'Credit/Debit Card'
mask = df['mode'] == 'Credit/Debit Card'
target_indices = df[mask].index.tolist()

# 2. Shuffle the indices randomly
np.random.seed(42)
np.random.shuffle(target_indices)

# 3. Split the list in half
midpoint = len(target_indices) // 2
zeros_indices = target_indices[:midpoint]
ones_indices = target_indices[midpoint:]

# 4. Assign the new values
df.loc[zeros_indices, 'mode'] = 0
df.loc[ones_indices, 'mode'] = 1

# 5. Convert the column to numeric (since it's now all 0s and 1s)
df['mode'] = df['mode'].astype(int)
```

```python
# 1. Identify the rows that have 'Credit/Debit Card'
mask = df['gender'] == 'Non-binary'
target_indices = df[mask].index.tolist()

# 2. Shuffle the indices randomly
np.random.seed(42)
np.random.shuffle(target_indices)

# 3. Split the list in half
midpoint = len(target_indices) // 2
zeros_indices = target_indices[:midpoint]
ones_indices = target_indices[midpoint:]

# 4. Assign the new values
df.loc[zeros_indices, 'gender'] = 0
df.loc[ones_indices, 'gender'] = 1

# 5. Convert the column to numeric (since it's now all 0s and 1s)
df['gender'] = df['gender'].astype(int)
```

```python
df['monthly_income'] = df['monthly_income']*2
```

```python
df['stationary'] = df['stationary']*1.5
df['food'] = df['food']*1.25
df['shoping'] = df['shoping']*2.25
df['friends'] = df['friends']*6
df['entertainment'] = df['entertainment']*2
df['personal'] = df['personal']*10
df['others'] = df['others']*2
```

```python
df['spent'] = df['food'] + df['entertainment'] + df['friends'] + df['others'] + df['personal'] + df['shoping'] +df['stationary']
```
